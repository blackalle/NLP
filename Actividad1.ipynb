```python
#importar librerias necesarias para el desarrollo de la actividad de nlp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import spacy
import nltk
```

```python
# leer el archico hotel.csv dentro de la carpeta data
hotel = pd.read_csv('data/hotel.csv')
hotel
```

```python
#normalizar los textos con la funcion normalize_document de mis_funciones.py
from mis_funciones import normalize_document
norm_hotel=[]

for document in hotel['text']:
    norm_hotel.append(normalize_document(document))

norm_hotel
```

```python
#obtener matrices bow y tf-idf utilizando los 20 tokens mas frecuentes
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

#bow con los 20 tokens mas frecuentes
cv = CountVectorizer(max_features=20)
bow = cv.fit_transform(norm_hotel).toarray()

#imprima las matrices en pantalla de forma fácilmente legible
bow_df = pd.DataFrame(bow, columns=cv.get_feature_names_out())
bow_df

```

```python
#tf-idf
tfidf = TfidfVectorizer(max_features=20)
tfidf_matrix = tfidf.fit_transform(norm_hotel).toarray()
tfidf_df = pd.DataFrame(np.round(tfidf_matrix, 2), columns=tfidf.get_feature_names_out())
tfidf_df
```

```python
#leer fichero esp.txt de la carpeta data
esp = pd.read_csv('data/esp.txt', sep='\t', header=None)
esp
```

```python
from spacy import displacy

# Cargar el modelo de lenguaje en español
nlp_es = spacy.load('es_core_news_lg')

# Iterar sobre todas las noticias en el DataFrame
for index, row in esp.iterrows():
    texto_noticia = row[0]
    doc = nlp_es(texto_noticia)
    
    # Visualizar las entidades con displacy
    displacy.render(doc, style="ent", jupyter=True)
```

```python
#leer fichero eng.txt de la carpeta data
eng = pd.read_csv('data/eng.txt', sep='\t', header=None)
eng
```

```python
# Cargar el modelo de lenguaje en inglés
nlp_en = spacy.load('en_core_web_lg')
# Iterar sobre todas las noticias en el DataFrame
for index, row in eng.iterrows():
    texto_noticia = row[0]
    doc = nlp_en(texto_noticia)
    
    # Visualizar las entidades con displacy
    displacy.render(doc, style="ent", jupyter=True)
```

#### El modelo en inglés es más completo ya que tiene mejor clasificadas las entidades del texto. El modelo en español llama MISC (MISCELÁNEA) a muchas entidades probablemente porque es una traducción del inglés. Traducciones como Thorin Oakenshield a Thorin Escudo de Roble o el uso en inglés de la apóstrofe en Bilbo's journey hacen que el modelo en inglés coja a esos personajes como entidades person y no MISC. 

#### Además, el modelo inglés tiene las entidades de language, cardinal o date lo cual puede ser muy interesante para los análisis nlp que se pueden hacer en el mundo profesional

#### En un entorno profesional, se podría entrenar al modelo con ejemplos como que Thorin Escudo de Roble sea una persona, que Tierra Media sea un lugar y demás para ayudar al modelo en nuestro proyecto específico acerca de El Hobbit.
