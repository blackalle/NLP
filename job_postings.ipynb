```python
#importar librerias necesarias para el desarrollo de la actividad de nlp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import spacy
import nltk

from sklearn.decomposition import LatentDirichletAllocation

nlp = spacy.load("es_core_news_lg")
```

```python
# leer el archico hotel.csv dentro de la carpeta data
job = pd.read_csv('job_postings.csv')
job.head()
```

```python
from nltk.tokenize import WordPunctTokenizer
from nltk.corpus import stopwords

wpt = WordPunctTokenizer()
stop_words_eng = set(stopwords.words('english'))

def normalize_doc(doc):
    doc = doc.replace("!", "").replace("¡", "").replace(",", "").replace(".", "").replace(";", "").replace("-", "").replace(":", "").replace("#", "").replace("?", "")
    doc = doc.lower()
    doc = doc.strip()
    # tokenize document
    tokens = wpt.tokenize(doc)
    # filter stopwords out of document
    filtered_tokens = [token for token in tokens if token not in stop_words_eng]
    # re-create document from filtered tokens
    doc = ' '.join(filtered_tokens)
    return doc
```

```python
#normalizar los textos con la funcion normalize_document de mis_funciones.py
norm_job=[]

for document in job['description']:
    norm_job.append(normalize_doc(document))

#ver solo los primeros 20 documentos normalizados
norm_job[:20]
```

```python
#convertir norm_spam en una matriz tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(norm_job)

tfidf_array = tfidf.toarray()
tfidf_array.shape
```

```python
# Visualización de la matriz

# Obtenemos el vocabulario para poner las etiquetas de las columnaas
vocab = tfidf_vectorizer.get_feature_names_out()

print("Palabras en el vocabulario: ", len(vocab))
print(vocab)
```

```python
# Y construimos un dataframe para mostrar el resultado: por cada documento las ocurrencias de cada token
pd.DataFrame(tfidf.toarray(), columns=vocab)
```

```python
# Creación de un objeto de clase LDA con sus componentes
lda_model = LatentDirichletAllocation(n_components = 4, max_iter = 20, random_state = 20)

# Extracción de los tópicos
X_topics = lda_model.fit_transform(tfidf)

topic_words = lda_model.components_
topic_words
```

```python
#  Número de palabras a extraer de cada tópico

n_top_words = 5

for i, topic_dist in enumerate(topic_words):
    
    sorted_topic_dist = np.argsort(topic_dist)
    
    topic_words = np.array(vocab)[sorted_topic_dist]
    
    topic_words = topic_words[:-n_top_words:-1]
    print ("Tópico", str(i+1), topic_words)
```
